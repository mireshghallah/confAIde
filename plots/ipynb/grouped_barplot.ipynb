```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
data_overall = pd.read_csv('people_tier_1_2ab_3_4_report - tier_summary_df.csv')
data_overall_melt = pd.melt(data_overall, id_vars =['Model'], value_vars =['Tier 1', 'Tier 2.a (n=5)','Tier 2.a (martins)','Tier 2.b'],var_name='Tier', value_name='Corr')

data_overall_melt 
```

```python
# set plot style: grey grid in the background:
sns.set(style="darkgrid") #darkgrid
sns.set(font_scale=1.3)

# Set the figure size
plt.figure(figsize=(16,8))

# grouped barplot
sns.barplot(x="Model", y="Corr", hue="Tier", data=data_overall_melt , ci=None)

plt.savefig('overview.pdf', transparent=False, bbox_inches='tight')
```
